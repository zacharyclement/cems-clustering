```python
import pandas as pd
import glob
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('2015cemsTest.csv')
```

```python
len(df['FACILITY_NAME'].unique())
```

```python
df['Gen Unit'] = df['FACILITY_NAME'].astype(str) + df['UNITID'].astype(str)
```

```python
df['OP_HOUR'] = df['OP_HOUR'].astype(str)
```

```python
df.dtypes
```

```python
df['Time'] = df['OP_DATE'] + '-' + df['OP_HOUR']
```

```python
df['Time'] = pd.to_datetime(df['Time'], format='%m-%d-%Y-%H')
```

```python
#replace nan with 0
df['GLOAD (MW)'].fillna(0, inplace=True)
```
